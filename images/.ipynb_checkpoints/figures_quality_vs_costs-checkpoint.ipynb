```python
from scipy import optimize,arange
from numpy import array
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
range_c = arange(0.0,1.1,0.05)


range_quality_1 = [1-c for c in range_c]

plt.clf()
 
plt.plot(range_c, range_c,'-', color = 'r', linewidth = 2)
plt.plot(range_c, range_quality_1,'ro', color = 'b', markersize = 8)



plt.title("well managed hospitals",fontsize = 25)
plt.xlabel("costs",fontsize = 25)
plt.ylabel("quality",fontsize = 25)

plt.xticks((),[]) # we don't need "ticks" on the axes
plt.yticks((),[])


plt.xlim(0.0,1.0)
plt.ylim(0.0,1.0)
plt.savefig('quality1.png')
```

```python
range_quality_2 = [0.5*c+0.25 for c in range_c]

plt.clf()
 
plt.plot(range_c, range_c,'-', color = 'r', linewidth = 2)
plt.plot(range_c, range_quality_2,'ro', color = 'b', markersize = 8)



plt.title("proton beam therapy",fontsize = 25)
plt.xlabel("costs",fontsize = 25)
plt.ylabel("quality",fontsize = 25)

plt.xticks((),[]) # we don't need "ticks" on the axes
plt.yticks((),[])


plt.xlim(0.0,1.0)
plt.ylim(0.0,1.0)
plt.savefig('quality2.png')
```

```python
range_quality_3 = [1.5*c-0.25 for c in range_c]

plt.clf()
 
plt.plot(range_c, range_c,'-', color = 'r', linewidth = 2)
plt.plot(range_c, range_quality_3,'ro', color = 'b', markersize = 8)



plt.title("graffiti in IC",fontsize = 25)
plt.xlabel("cost",fontsize = 25)
plt.ylabel("quality",fontsize = 25)

plt.xticks((),[]) # we don't need "ticks" on the axes
plt.yticks((),[])


plt.xlim(0.0,1.0)
plt.ylim(0.0,1.0)
plt.savefig('quality3.png')

```
